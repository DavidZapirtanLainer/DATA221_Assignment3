```python
import pandas as pd
from sklearn.model_selection import train_test_split

kidney_disease_data_frame = pd.read_csv("D:/datasets/kidney_disease.csv")


#Including all columns except the last.
feature_matrix = kidney_disease_data_frame.drop("classification", axis=1)

#Making labels = last column
target_labels = kidney_disease_data_frame["classification"]

#Setting test_size to 0.3 so 30% of data is testing data
model_features_train, model_features_test, model_labels_train, model_labels_test = train_test_split(feature_matrix, target_labels, test_size= 0.3, random_state=221)

```

```python
# Why should we not train and test a model on the same data?
# Testing a model on the same data that it was trained with makes it overfit to that specific data.The model will have an unrealistic high accuracy as a result of memorizing data points rather than learning its patterns. Ultimately this will cause the model to fail when given unseen data, resulting in a high variance.
```

```python
# What's the purpose of the testing set?
# The purpose of the testing set is to see how the model performs when given unseen data points. It is very useful as one can compare the model's predictions of the testing set with its actual values and calculate metrics like  accuracy, precision, recall, and f1 score that can tell us how well the model performs.
```
