```python
import pandas as pd
import matplotlib.pyplot as plt

crime_file = pd.read_csv("D:/datasets/crime1.csv")

violent_crimes_per_population = crime_file.loc[:, "ViolentCrimesPerPop"]

#In this block we imported pandas and matplotlib for making a dataframe and plotting our graphs respectively.
```

```python
#Plotting the Histogram
plt.hist(violent_crimes_per_population, bins = 40, color = "skyblue", edgecolor = "black")
plt.xlabel('Violent Crimes per Population')
plt.ylabel('Frequency (Count)')
plt.title('Distribution of Different Levels of Violent Crimes per Population')
plt.show()
```

```python
# What does the histogram shows about how the data values are spread?
# The histogram seems to be a right skewed histogram with the bulk of values sitting on the lower end of the violent crimes per population statistic (0.0 - 0.5). Furthermore, looking at the histogram we can also see that the most common value for violent crimes per population is 1.0, this explains why the average was greater median.
```

```python
#Plotting the box plot
plt.boxplot(violent_crimes_per_population)
plt.title("Violent Crimes per Population across Communities ")
plt.xlabel("All Communities")
plt.ylabel("Violent Crimes per Population")
plt.show()
```

```python
# What does the boxplot tell us about the median?
# Inside the boxplot, there is an orange line, this is our median. Thus, we can see that the median sits roughly at 0.4 violent crimes per population.
```

```python
# Does the boxplot suggest the existence of outliers?
# No, the boxplot does not suggest there are any outliers in this dataset. In order for the boxplot to show the existence of outliers, there would have to be circles/dots outside of the whiskers of the box, which represent the upper and lower fences of the data set. Since there are no circles/dots outside of the boxplot's whiskers, we can conclude that there are no outliers.
```
