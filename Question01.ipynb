```python
import pandas as pd
crime_file = pd.read_csv("D:/datasets/crime1.csv")

violent_crimes_per_population = crime_file.loc[:, "ViolentCrimesPerPop"].agg(["mean", "median", "std", "min", "max"])

print(violent_crimes_per_population)


# 1. Compare the mean and median. Does the distribution look symmetric or skewed? Explain:
# The mean for this column is roughly 0.44 and the median is 0.39, what does this tell us?

# The median is the actual middle point of the data set, while the mean is the sum of all
# values divided by the number of total values. While this is good to find averages, when
# there are outliers, the mean can easily be skewed. Here we see that the mean is 0.05
# more to the right compared to the actual middle value. Hence, there are outliers that
# are values closer to 1.

# If there are extreme values (very large or very small), which statistic is more affected: mean
# or median? Explain why

# As expressed previously, the extremes values seem to be very large rather than very small. This
# affects the mean because it involves these extreme values into the sum before dividing by the
# number of total values. The median is never affected by extreme values as it is always just the
# middle value of a dataset.
```
